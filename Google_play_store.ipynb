```python
def content_rating(text):
  if text=="Everyone" or text=="Everyone 10+":
    return "Everyone"
  elif(text=="Mature 17+"):
    return "Mature"
  elif(text=="Adults only 18+"):
    return "Adults"
  else:
    return text


df["Content Rating"]=df["Content Rating"].apply(content_rating)
```
